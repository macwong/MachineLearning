# Week 2 - Linear Regression with Multiple Variables

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("ex1data2.txt", header=None)
```

```python
data[data[0] > 4000].head(50)
```

```python
X = data[[0, 1]]
y = data[2]
m = len(y) 
```

```python
mean = X.mean()
mean
```

```python
std = np.sqrt(np.sum((X - mean) ** 2) / (m - 1))
std
```

```python
X_scaled = (X - mean) / std
```

```python
X_scaled
```

```python
X_scaled.insert(0, "bias", np.ones(m))
```

```python
X_scaled.columns = ['bias', 'size', 'bedrooms']
X_scaled.head()
```

```python
theta = np.zeros(len(X_scaled.columns))
```

```python
theta
```

```python
# Cost function
h = X_scaled.dot(theta)
J = (1 / (2 * m)) * (np.sum(h - y) ** 2)
J
```

```python
print((h - y).shape)
print(X_scaled.shape)
print(np.sum((h - y).T.dot(X_scaled)))
print(X_scaled.T.dot(h - y))
print(theta)
```

```python
# Gradient Descent
alpha = 0.1
iterations = 1000
theta = np.zeros(len(X_scaled.columns))

for i in range(iterations):
    h = X_scaled.dot(theta)
    theta = theta - alpha * (1 / m) * X_scaled.T.dot(h - y)
#     J = (1 / (2 * m)) * (np.sum(h - y) ** 2)
    J = (1 / (2 * m)) * (h - y).T.dot(h-y)
    
    print(theta)
    print(J)
```

```python
X_predict = [4000, 4]
X_predict_scaled = ((X_predict - mean) / std).to_numpy()
X_predict_scaled = np.insert(X_predict_scaled, 0, 1, axis=0)
predict = np.dot(X_predict_scaled, theta)
```
