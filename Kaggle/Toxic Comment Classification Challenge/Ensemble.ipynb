```python
import numpy as np, pandas as pd

f_lstm = 'LSTM baseline - submission.csv'
f_nbsvm = 'Baseline - submission.csv'
```

```python
p_lstm = pd.read_csv(f_lstm)
p_nbsvm = pd.read_csv(f_nbsvm)
```

```python
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
p_res = p_lstm.copy()
p_res[label_cols] = (p_nbsvm[label_cols] + p_lstm[label_cols]) / 2
```

```python
p_res.to_csv('Ensemble - submission.csv', index=False)
```

```python
p_res2 = p_lstm.copy()

for rowIndex in range(len(p_res2)): 
    for colIndex in label_cols:
        val_nbsvm = p_nbsvm[colIndex][rowIndex] 
        val_lstm = p_lstm[colIndex][rowIndex]
        
        val = (val_nbsvm + val_lstm) / 2
        
        if val_nbsvm > 0.5 and val_lstm > 0.5:
            val = np.maximum(val_nbsvm, val_lstm)
        elif val_nbsvm < 0.2 and val_lstm < 0.2:
            val = np.minimum(val_nbsvm, val_lstm)
            
        p_res2.set_value(rowIndex, colIndex, val)
    
print(p_res2.shape)
p_res2.head()
```

```python
p_res2.to_csv('Ensemble - submission.csv', index=False)
```
